```python
import requests as rs
```

```python
from bs4 import BeautifulSoup as bs
```

```python
import pandas as pd
```

```python
url="https://quotes.toscrape.com/page/1/"
```

```python
response=rs.get(url)
print(response)
```

```python
response=response.content
print(response)
```

```python
soup=bs(response,"html.parser")

```

```python
container=soup.find('div',class_='container')
print(container)
```

```python
contents=container.find_all('div',class_='quote')
for i in contents:
    print("the contents are:",i)
```

```python
for content in contents:
   
    quotes=content.find(class_='text')
    quotes=quotes.text
    print(quotes)
```

```python
for content in contents:
    author=content.find(class_='author')
    author=author.text
    print(author)
```

```python
contents=container.find_all('span',class_='tag-item')
tags=[]
for content in contents:
#     print("CON:",content)
    tag=content.find('a',class_='tag')
    tag=tag.text
    tags.append(tag)
print(tags)

    
```

```python
result=[]
for i in range(1,3):
# for i in tags:
    url=f"http://quotes.toscrape.com/tag/{i}/"
    url=f"http://quotes.toscrape.com/page/{i}/"
    response=rs.get(url)
    response=response.content
    soup=bs(response,'html.parser')
    container=soup.find('div',class_='container')
    contents=container.find_all('div',class_='quote')

    
    for content in contents:
        quotes=content.find(class_='text')
        quotes=quotes.text
#         quotes_lst.append(quotes)
        author=content.find(class_='author')
        name=author.text
#         name_lst.append(author)
        tag=content.find('a',class_='tag')
        tag=tag.text
#         tags_lst.append(tag)
        result.append([quotes,name,tag])

           
```

```python
data=pd.DataFrame(result,columns=["Quotes","Name of the Author","Name of the tag"])
```
